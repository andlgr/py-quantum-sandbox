```python
from qiskit import transpile
from qiskit import QuantumCircuit
from qiskit.quantum_info import Operator
from qiskit.visualization import plot_distribution
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector

qc = QuantumCircuit(3)

qc.h(1)
qc.h(2)

qc.ccx(2, 1, 0)
display(qc.draw('mpl'))

display(Statevector(qc).draw('latex'))
display(Operator(qc).draw('latex'))

qc.measure_all()

# Run simulation
simulator = AerSimulator()
qc_t = transpile(qc, simulator)
result = simulator.run(qc_t, shots=100000).result()

counts = result.get_counts()
plot_distribution(counts)
```

```python
from qiskit import QuantumRegister

n = 3
qr_x = QuantumRegister(n, name='x')
qr_y = QuantumRegister(1, name='y')

# Implement reversible 3-qubit AND
qc_and3 = QuantumCircuit(qr_y, qr_x)
qc_and3.h(qr_x)
qc_and3.mcx(qr_x,qr_y)


qc_and3.draw('mpl')

```

```python
qc_and3.measure_all()

# Run simulation
simulator = AerSimulator()
qc_t = transpile(qc_and3, simulator)
result = simulator.run(qc_t, shots=1000000).result()

counts = result.get_counts()
plot_distribution(counts)
```
